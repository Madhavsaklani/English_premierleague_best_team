```python
import pandas as pd
team_df = pd.read_csv("../data/processed/matches_cleaned.csv")
summary = pd.read_csv("../data/processed/team_summary.csv")

summary.head()

```

```python
team_df.head()
```

```python
# Do teams with more shots always score more goals?
```

```python
attack_summary = (
    team_df
    .groupby('Team', as_index=False)
    .agg({
        'Shots': 'sum',
        'GoalsFor': 'sum',
        'Points': 'sum'
    })
)

# derived metrics
attack_summary['GoalConversion'] = attack_summary['GoalsFor'] / attack_summary['Shots']
```

```python
attack_summary\
    .sort_values('GoalsFor', ascending=False)\
    .head(20)
```

```python
attack_summary.to_csv('../data/insights/insight_01_attacking_efficiency.csv', index=False)
```

```python
# Thus we can coclude that teams with more shots always have a better chance of goal conversion.
# But not true in every case with more shot there are always a chance miss the shot. 
# like wise here we can see Newcastle United has the highest goal conversion ratio.
# has more goal than Manchester United.
```

```python
#Does strong defense contribute more to league success than attack?
```

```python
defence_summary = (
    team_df
    .groupby('Team', as_index=False)
    .agg({
        'CleanSheet': 'sum',
        'GoalsAgainst': 'sum',
        'Points': 'sum'
    })
)

defence_summary\
    .sort_values('Points', ascending=False)\
    .head(20)
```

```python
# thus we can conclude that Strong defensive performance (clean sheets and low goals conceded) is essential for maintaining league consistency, 
# but it does not alone guarantee the league title. 
# Elite teams combine solid defense with superior attacking efficiency.
```

```python
defence_summary.to_csv('../data/insights/insight_02_defensive_efficiency.csv', index=False)
```

```python
#overperforming teams 
```

```python
overperform_df = (
    team_df
    .groupby('Team', as_index=False)
    .agg({
        'GoalsFor': 'sum',
        'Points': 'sum'
    })
)
```

```python
overperform_df['PointsPerGoal'] = (
    overperform_df['Points'] / overperform_df['GoalsFor']
)
```

```python
overperform_df\
    .sort_values('PointsPerGoal', ascending=False)\
    .head(20)
```

```python
#Football is not about scoring more goals, itâ€™s about scoring goals at the right time.
```

```python
overperform_df.to_csv('../data/insights/insight_03_overperformers.csv', index=False)
```
