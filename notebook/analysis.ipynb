```python
home = df[[
    'Date','HomeTeam','AwayTeam',
    'FTHG','FTAG','FTR',
    'HS','HST','HC','HF','HY','HR'
]].copy()

home.columns = [
    'Date','Team','Opponent',
    'GoalsFor','GoalsAgainst','Result',
    'Shots','ShotsOnTarget','Corners','Fouls','Yellow','Red'
]

home['Venue'] = 'Home'
home['Points'] = home['Result'].map({'H':3,'D':1,'A':0})
```

```python
away = df[[
    'Date','AwayTeam','HomeTeam',
    'FTAG','FTHG','FTR',
    'AS','AST','AC','AF','AY','AR'
]].copy()

away.columns = [
    'Date','Team','Opponent',
    'GoalsFor','GoalsAgainst','Result',
    'Shots','ShotsOnTarget','Corners','Fouls','Yellow','Red'
]

away['Venue'] = 'Away'
away['Points'] = away['Result'].map({'A':3,'D':1,'H':0})
```

```python
team_df = pd.concat([home, away], ignore_index=True)
```

```python
team_df['ShotAccuracy'] = team_df['ShotsOnTarget'] / team_df['Shots']
team_df['GoalConversion'] = team_df['GoalsFor'] / team_df['ShotsOnTarget']
team_df['CleanSheet'] = (team_df['GoalsAgainst'] == 0).astype(int)
```

```python
team_df.info()
team_df.isna().sum().sort_values(ascending=False).head(10)
```

```python
team_df['ShotAccuracy'] = team_df['ShotsOnTarget'] / team_df['Shots']
team_df['GoalConversion'] = team_df['GoalsFor'] / team_df['ShotsOnTarget']

team_df.replace([float('inf'), -float('inf')], 0, inplace=True)
team_df.fillna(0, inplace=True)
```

```python
team_df['Result'].value_counts()
```

```python
summary = team_df.groupby('Team').agg(
    Matches=('Team','count'),
    Points=('Points','sum'),
    Wins=('Points', lambda x: (x==3).sum()),
    Draws=('Points', lambda x: (x==1).sum()),
    Losses=('Points', lambda x: (x==0).sum()),
    GoalsFor=('GoalsFor','sum'),
    GoalsAgainst=('GoalsAgainst','sum'),
    Shots=('Shots','sum'),
    ShotsOnTarget=('ShotsOnTarget','sum'),
    Corners=('Corners','sum'),
    CleanSheets=('CleanSheet','sum')
).reset_index()

summary['GoalDiff'] = summary['GoalsFor'] - summary['GoalsAgainst']
summary = summary.sort_values(['Points','GoalDiff'], ascending=False)
```

```python
summary.to_csv("../data/processed/team_summary.csv", index=False)
```
